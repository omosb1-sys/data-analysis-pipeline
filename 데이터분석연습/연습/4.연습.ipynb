```python
#아우디가 생산한 차량중에 고속도로 연비가 높은 자동차 모델 알아보기
# query() sort_values(,acending=False) head() 힌트
mpg.query('manufacturer == "audi"').sort_values('hwy', ascending=False).head()
```

```python
#4. 데이터 복사본을 만들어 작업해보기
# 4-1 도시연비와 고속도로 연비를 더한 합산 연비 컬럼 구하기(assign함수)
# 4-2 평균 연비 변수 추가하기
# 4-3 평균 연비가 가장 높은 3종의 데이터만 추출하기

mpg_new = mpg.copy()
```

```python
mpg_new
```

```python
mpg_new = mpg_new.assign(total = mpg_new['cty'] + mpg_new['hwy'])
mpg_new
```

```python
mpg_new = mpg_new.assign(mean = mpg_new['total'] / 2)
mpg_new
```

```python
mpg_new.sort_values('mean', ascending=False).head()
```

```python
mpg.assign(total = lambda x: x['cty'] + x['hwy'],
           mean = lambda x: x['total'] / 2
           ).sort_values('mean',ascending=False
                         ).head()
```
