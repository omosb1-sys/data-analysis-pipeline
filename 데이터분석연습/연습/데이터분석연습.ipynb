```python
import pandas as pd
#1. 데이터 불러오기
transcript = pd.read_csv("/Users/sebokoh/Downloads/transcript 3.csv")
profile = pd.read_csv("/Users/sebokoh/Downloads/profile.csv")
```

```python
# 2. 거래 이벤트만 필터링
transactions = transcript[transcript['event'] == 'transaction']
```

```python
#3. amount 컬럼 추출 
def safe_amount(v):
    if isinstance(v, dict):
        return v.get('amount', 0)
    return 0

transactions['amount'] = transactions['value'].apply(safe_amount)
```
