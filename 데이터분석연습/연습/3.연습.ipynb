```python
data["Fare_group"] = pd.qcut(data["Fare"], 3,
                             labels = ["L", "M","H"])
```

```python
data[["PassengerId", "Fare", "Fare_group"]]
```

```python
data["Fare_level"] = data.groupby("PassengerId")["Fare_group"].\
    transform(lambda x: pd.qcut(x.astype(float), 3, labels=["L", "M", "H"], duplicates="drop").astype(str))
    
    
```
